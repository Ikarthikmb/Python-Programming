```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

df=pd.read_csv('DataFiles/Income.csv')
y=df.values[3]
x=df.columns
```

```python
years=[]
for i in range (2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])

plt.xlabel('Years')
plt.ylabel('Income')
plt.title('Income of Arizona')
plt.plot(years,income)
plt.grid()
plt.show
```

```python
width=1/1.5
plt.bar(years,income,width)
plt.title('Income of Arizona')
plt.xlabel('Years')
plt.ylabel('Income')
plt.savefig('Figures/ArizonaIncomeBarGraph.png')
```
