## Problem 4:
## Statement:
Delete a row from .csv file

```python
def deleteRowCSV(filename,rowindex):
    df=readCSV(filename)
    df=df.drop(rowindex)
    df.to_csv(filename,index=False)
    return df

deleteRowCSV('DataFiles/Income.csv',4)
```
